```python
import pandas as pd
import matplotlib.pyplot as plt
# Plot a histogram for the 'views' column
plt.figure(figsize=(10, 6))
plt.hist(df['video views'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Video Views on YouTube')
plt.xlabel('Number of Views')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Scatter Plot of Views vs Likes
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10, 6))
plt.scatter(df['video views'], df['subscribers'], color='green', alpha=0.5)
plt.title('video views vs subscribers')
plt.xlabel('Number of video views')
plt.ylabel('Number of subscribers')
plt.grid(True)
plt.show()
```
